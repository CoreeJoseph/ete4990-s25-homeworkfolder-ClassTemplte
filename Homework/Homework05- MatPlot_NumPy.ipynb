#Part 1 Mat Plot Lib

Using Matplotlib create the following graphs


````python
# Inports Here
import matplotlib.pyplot as plt
import numpy as np
#from mpl_toolkits.mplot3d import Axes3d
````

````python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequecy')
plt.show()
````

````python
# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````

````python
# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='green', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

````python
# Pie Plot
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Plot')
plt.show()
````

````python
# Area Plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, color="skyblue", alpha=0.5, label='Sine')
plt.fill_between(x, y2, color="orange", alpha=0.5, label='Cosine')
plt.title('Area Plot')
plt.legend()
plt.show()
````
